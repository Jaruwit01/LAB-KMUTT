```python
import numpy as np
import random
def objective(x, d):
    ans=0
    multi=1
    for j in range(d):
        multi *= np.cos(x[j] / np.sqrt(j + 1))
        
    for i in range(d):
        ans += ((x[i] * x[i]) / 4000)
        #print(i, ans)
    return ans - multi + 1

x=[0 for i in range(70)]


print(objective(x, 7))
print(objective(x, 70))
```
