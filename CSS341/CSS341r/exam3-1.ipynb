```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#3.1 group day of week
data=pd.read_csv('exan/data/US-accidents.csv')
data['Start_Time']=pd.to_datetime(data['Start_Time'], unit='ns')
data['Start_Time']=data['Start_Time'].dt.day_name()#show day of week
temp1 = data.groupby(['Start_Time']).count()['ID']#group day of week
temp1
```

```python
#3.1 group by time of day
data=pd.read_csv('exan/data/US-accidents.csv')
data['Start_Time']=pd.to_datetime(data['Start_Time'], unit='ns')
data['Start_Time']=data['Start_Time'].dt.hour//6
t=['night' ,'morning' , 'afternoon', 'evening']
data['Start_Time']=data['Start_Time'].apply(lambda x: t[x])
temp2 = data.groupby(['Start_Time']).count()['ID']
temp2
```

```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return ('{v:d}'.format(v=val))
    return my_autopct
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,10)) #ax1,ax2 refer to your two pies
mylabels1 = temp1.index
mylabels2 = temp2.index
ax1.pie(temp1,labels = mylabels1,autopct = make_autopct(temp1)) #plot first pie
ax2.pie(temp2,labels = mylabels2,autopct = make_autopct(temp2)) #plot second pie
pass
```
