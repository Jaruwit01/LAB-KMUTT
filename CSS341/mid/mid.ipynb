```python
#9
file = "data2/tx201"
tx = pd.read_csv(file + "0.csv")
for i in range(1,6):
  path = file + str(i) + ".csv"
  tx_read = pd.read_csv(path)
  tx = tx.merge(tx_read, how = 'outer')
display(tx)
```

```python
customer = pd.read_csv("data2/customers.csv")
```

```python
#10
no_discount = tx[pd.isnull(tx['discount'])]
have_discount = pd.concat([tx, no_discount], ignore_index=True).drop_duplicates(subset=['customer_id', 'transaction_date', 'ticket_price', 'discount', 'product'], keep=False)
have_discount['discount'] = have_discount['discount'].apply(lambda x: x[:-1]).astype(float)
#have_discount['discount']
have_discount['ticket_price'] = have_discount['ticket_price'] - (have_discount['ticket_price'] * have_discount['discount'] / 100).astype(float)
all = no_discount.merge(have_discount, how = 'outer')
#all = all.merge(customer, how='inner', on='customer_id')
all_merge = all.groupby(['customer_id']).sum(['ticket_price']).merge(customer, how='inner', on='customer_id')
sorted = all_merge.sort_values(by = ['ticket_price'], ascending = False)
print("10.1")
display(sorted[['customer_id', 'ticket_price','address_zip', 'address_state']])
print("10.2\n","top20")
display(sorted[:20])
print("bottom20")
display(sorted[-20:])
```
